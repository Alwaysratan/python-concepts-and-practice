```python

# What are Lists?


# A list is an ordered, mutable collection that can store
# multiple items of different data types.

numbers = [10, 20, 30, 40]

print(numbers)
print(type(numbers))
```

```python

# Creating Lists


# Lists can store integers, strings, floats, booleans,
# or even other lists.

numbers = [10, 20, 30]

names = ["Ratan", "Rahul", "Amit"]

mixed = [10, "Python", 3.14, True]

nested = [[1, 2], [3, 4]]

print(numbers)
print(names)
print(mixed)
print(nested)
```

```python

# Indexing


# Indexing is used to access individual elements of a list.
# Positive indexing starts from 0, while negative indexing
# starts from -1.

fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[0])
print(fruits[2])

print(fruits[-1])
print(fruits[-2])
```

```python

# Slicing


# Slicing extracts a portion of a list using start:stop:step.

numbers = [10,20,30,40,50,60]

print(numbers[1:4])

print(numbers[:3])

print(numbers[3:])

print(numbers[::2])

print(numbers[::-1])
```

```python
#iteration using for loop
a=["ironman","thor","cap","halk"]
for i in a:
    print(i)
```

```python
#iteration using for loop with range and length function
for i in range(len(a)):
    print(a[i])
```

```python
#iteration using while
i=0
while i<len(a):
    print(a[i])
    i+=1
```

```python
#using short hand for loop
[print(i) for i in a]
```

```python
#to find the length of a list
print(len(a))
```

```python
#to count an occurence of a particular element
print(a.count("thor"))
```

```python

# append()


# append() adds a single element to the end of the list.

fruits = ["Apple", "Banana"]

fruits.append("Mango")

print(fruits)
```

```python

# extend()


# extend() adds multiple elements from another iterable.

fruits = ["Apple", "Banana"]

fruits.extend(["Mango", "Orange"])

print(fruits)
```

```python

# insert()


# insert() inserts an element at a specified position.

fruits = ["Apple", "Banana"]

fruits.insert(1, "Mango")

print(fruits)
```

```python

# remove()


# remove() deletes the first occurrence of a specified value.

fruits = ["Apple", "Banana", "Mango"]

fruits.remove("Banana")

print(fruits)
```

```python

# pop()


# pop() removes and returns an element by index.
# If no index is given, it removes the last element.

numbers = [10,20,30,40]

print(numbers.pop())

print(numbers)

print(numbers.pop(1))

print(numbers)
```

```python

# clear()


# clear() removes all elements from the list.

numbers = [10,20,30]

numbers.clear()

print(numbers)
```

```python

# copy()


# copy() creates a shallow copy of a list.

numbers = [10,20,30]

new_numbers = numbers.copy()

print(new_numbers)
```

```python

# index()


# index() returns the position of the first occurrence
# of a specified element.

fruits = ["Apple","Banana","Mango"]

print(fruits.index("Banana"))
```

```python

# reverse()


# reverse() reverses the order of elements in the list.

numbers = [10,20,30,40]

numbers.reverse()

print(numbers)
```

```python

# sort()


# sort() sorts the list in ascending order by default.
# Use reverse=True for descending order.

numbers = [40,10,30,20]

numbers.sort()

print(numbers)

numbers.sort(reverse=True)

print(numbers)
```

```python
#list comprehension
l1=[30,40,5,6]
l2=[]
for i in l1:
    l2.append(i)
print(l2)
l3=[i for i in l1 if i>6]
print(l3)
```
