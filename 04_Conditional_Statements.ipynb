```python

# What are Conditional Statements?


# Conditional statements allow a program to make decisions
# based on whether a condition is True or False.

age = 20

if age >= 18:
    print("You are eligible to vote.")
```

```python

# if Statement


marks = 80

if marks >= 40:
    print("Pass")
```

```python
salary = 50000

if salary > 30000:
    print("Eligible for Loan")

if salary > 45000:
    print("Premium Customer")
```

```python

# if-else Statement


num = 15

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
age = 16

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

```python

# if-elif-else Statement


marks = 72

if marks >= 90:
    print("Grade A+")

elif marks >= 80:
    print("Grade A")

elif marks >= 70:
    print("Grade B")

elif marks >= 60:
    print("Grade C")

else:
    print("Fail")
```

```python

# Nested if Statement


age = 22
citizen = True

if age >= 18:

    if citizen:
        print("Eligible to Vote")

    else:
        print("Not a Citizen")

else:
    print("Under Age")
```

```python

# Short Hand if


age = 20

if age >= 18: print("Adult")
```

```python

# Short Hand if-else


age = 16

print("Adult") if age >= 18 else print("Minor")
```

```python
age = 25
salary = 60000

if age >= 18 and salary >= 50000:
    print("Eligible")
```

```python
marks = 35

if marks >= 40 or marks >= 35:
    print("Eligible for Grace Marks")
```

```python
logged_in = False

if not logged_in:
    print("Please Login")
```

```python

# Practice Questions


# 1. Check whether a number is positive, negative, or zero.
num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")
```

```python
# 2. Check whether a number is even or odd.


num = int(input("\nEnter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
# 3. Find the largest of two numbers.


a = int(input("\nEnter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Largest Number =", a)
else:
    print("Largest Number =", b)
```

```python
# 4. Find the largest of three numbers.


a = int(input("\nEnter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print("Largest Number =", a)
elif b >= a and b >= c:
    print("Largest Number =", b)
else:
    print("Largest Number =", c)
```

```python
#5. Check whether a year is a leap year.


year = int(input("\nEnter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# Check whether a character is a vowel or consonant.


char = input("\nEnter a character: ").lower()

if char in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```
