```python

# What are Loops?


# Loops are used to execute a block of code repeatedly
# until a condition becomes False or for each item in a sequence.

for i in range(5):
    print(i)
```

```python

# while Loop


count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
#1.Sum of First 10 Natural Numbers
total = 0
num = 1

while num <= 10:
    total += num
    num += 1

print("Sum =", total)
```

```python
#wtp to check if a number is prime or not
num=int(input("enter a number :"))
if num<=1:
    print("not prime ")
    
else:
    for i in range(2,num):
        if num%i==0:
            print("number is not prime number")
            break
    else:
            print("it is a prime")
```

```python
#palindrom number
num=int(input("enter the number"))
temp=num
rev=0
while num>0:
    dig=num%10
    rev=rev*10+dig
    num=num//10
if rev==temp:
    print("it is palindrome")
else:
    print("it is not palindrome")
```
