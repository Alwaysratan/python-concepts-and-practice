```python

# File Modes


# "r"  -> Read (Default)

# "w"  -> Write (Creates new file if not exists)

# "a"  -> Append

# "x"  -> Create

# "rb" -> Read Binary

# "wb" -> Write Binary
```

```python
#reading file
file=open("report01.txt","r")
data=file.read()
file.close() #always close the file when work done with file
```

```python
#with keyword
#with keyword automatically close the file we don't need to write manually write file.close()
with open("report.txt","r") as f:
    data=f.read() #read all the file content
    print("file data",data)
```

```python
#readline by line file
with open("report01.txt","r") as f:
    line1=f.readline()
    line2=f.readline()
    line3=f.readline()
    print("line1",line1)
    print("line2",line2)
    print("line3",line3) #note : after reading all the line then you put any function like .read() , .readline() etc it gives empty file
```

```python
#read all lines
with open("report01.txt","r") as f:
    lines=f.readlines() #it give all the line from the file in form of list of string
    print(lines) 
```

```python
#2. write a program to read only the first line of report01.text
with open("report01.txt","r") as f:
    line1=f.readline()
    print(line1)
```

```python
#3.print how many lines are present in report01.txt
with open("report01.txt","r") as f:
    lines=f.readlines()
    print("numbers of line",len(lines))
```

```python

# seek()


with open("simple.txt", "r") as file:

    file.read(10)

    file.seek(0)

    print(file.read())
```

```python

# tell()


with open("simple.txt", "r") as file:

    print(file.tell())

    file.read(5)

    print(file.tell())
```
