```python

# Creating Dictionary


# A dictionary is an unordered (ordered from Python 3.7+),
# mutable collection that stores data in key-value pairs.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

print(student)
print(type(student))
```

```python

# keys()


# keys() returns all the keys in the dictionary.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

print(student.keys())
```

```python

# values()


# values() returns all the values in the dictionary.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

print(student.values())
```

```python

# items()


# items() returns all key-value pairs as tuples.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

print(student.items())
```

```python

# get()


# get() returns the value of the specified key.
# It returns None if the key does not exist.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

print(student.get("name"))
print(student.get("city"))
print(student.get("city", "Not Found"))
```

```python

# update()


# update() adds new key-value pairs or updates existing ones.

student = {
    "name": "Ratan",
    "age": 22
}

student.update({"age": 23})

student.update({"city": "Guwahati"})

print(student)
```

```python

# pop()


# pop() removes the specified key and returns its value.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

removed = student.pop("age")

print("Removed:", removed)

print(student)
```

```python

# popitem()


# popitem() removes and returns the last inserted key-value pair.

student = {
    "name": "Ratan",
    "age": 22,
    "course": "Python"
}

item = student.popitem()

print(item)

print(student)
```

```python

# clear()


# clear() removes all key-value pairs from the dictionary.

student = {
    "name": "Ratan",
    "age": 22
}

student.clear()

print(student)
```

```python

# copy()


# copy() creates a shallow copy of the dictionary.

student = {
    "name": "Ratan",
    "age": 22
}

new_student = student.copy()

print(new_student)
```

```python

# setdefault()


# setdefault() returns the value of the key.
# If the key does not exist, it adds the key with the given value.

student = {
    "name": "Ratan",
    "age": 22
}

print(student.setdefault("city", "Guwahati"))

print(student)

print(student.setdefault("name", "Rahul"))

print(student)
```

```python

# fromkeys()


# fromkeys() creates a new dictionary with the specified keys
# and a common default value.

keys = ("name", "age", "course")

student = dict.fromkeys(keys, "Not Available")

print(student)
```

```python
#practice questions
# Find the Student with the Highest Marks


students = {
    "Ratan": 85,
    "Rahul": 92,
    "Amit": 78,
    "Priya": 95
}

highest_marks = max(students.values())

for name, marks in students.items():
    if marks == highest_marks:
        print(name, ":", marks)
```

```python
# Count the Frequency of Each Character


text = "programming"

frequency = {}

for char in text:
    frequency[char] = frequency.get(char, 0) + 1

print(frequency)
```

```python
# Merge Two Dictionaries


dict1 = {
    "A": 10,
    "B": 20
}

dict2 = {
    "C": 30,
    "D": 40
}

dict1.update(dict2)

print(dict1)
```

```python
# Create a Nested Dictionary and Access an Inner Value


students = {
    "student1": {
        "name": "Ratan",
        "marks": 90
    },
    "student2": {
        "name": "Rahul",
        "marks": 85
    }
}

print(students["student1"]["marks"])
```

```python
# Reverse Keys and Values


student = {
    "A": 10,
    "B": 20,
    "C": 30
}

reverse_dict = {}

for key, value in student.items():
    reverse_dict[value] = key

print(reverse_dict)
```

```python
# Create a Dictionary Using Dictionary Comprehension


numbers = [1, 2, 3, 4, 5]

square = {num: num ** 2 for num in numbers}

print(square)
```

```python
# Create a Dictionary from Two Lists


keys = ["name", "age", "city"]

values = ["Ratan", 22, "Guwahati"]

student = dict(zip(keys, values))

print(student)
```

```python
# Count the Frequency of Words in a Sentence


sentence = "python is easy python is powerful"

words = sentence.split()
print(words)
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print(frequency)
```
