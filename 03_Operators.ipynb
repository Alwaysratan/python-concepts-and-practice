```python
#comparision opertor

a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#assignment oparetor

x = 10

# +=
x += 5
print(x)

# -=
x -= 3
print(x)

# *=
x *= 2
print(x)

# /=
x /= 4
print(x)

# //=
x //= 2
print(x)

# %=
x %= 3
print(x)

# **=
x **= 2
print(x)
```

```python
#bitwise operator
#compare the binary 
a = 5
b = 3

# AND
print(a & b)

# OR
print(a | b)

# XOR
print(a ^ b)

# NOT
print(~a)

# Left Shift
print(a << 1)

# Right Shift
```
