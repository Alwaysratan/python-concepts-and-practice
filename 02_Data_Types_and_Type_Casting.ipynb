```python

# Numeric Data Types


# Integer (int)
age = 21
print(age)
print(type(age))

# Float (float)
height = 5.8
print(height)
print(type(height))

# Complex (complex)
num = 3 + 4j
print(num)
print(type(num))
```

```python

# String


name = "Ratan"

print(name)
print(type(name))

# Accessing Characters

print(name[0])
print(name[-1])

# String Length

print(len(name))
```

```python

# Boolean


is_student = True
is_employee = False

print(is_student)
print(is_employee)

print(type(is_student))
```

```python

# type()


a = 10
b = 5.5
c = "Python"
d = True
e = [1, 2, 3]
f = (1, 2, 3)
g = {1, 2, 3}
h = {"name": "Ratan"}

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
```

```python

# int()


print(int(10.8))
print(int("25"))

# print(int("Hello"))    # Error
```

```python

# float()


print(float(10))
print(float("25.5"))

# print(float("Python"))   # Error
```

```python

# str()


print(str(100))
print(str(3.14))
print(str(True))
```

```python

# bool()


print(bool(1))
print(bool(0))

print(bool("Hello"))
print(bool(""))

print(bool([]))
print(bool([1, 2]))

print(bool({}))
print(bool({"name": "Ratan"}))
```

```python

# Implicit Type Casting


a = 10      # int
b = 2.5     # float

result = a + b

print(result)
print(type(result))
```

```python

# Explicit Type Casting


num = "100"

print(type(num))

num = int(num)

print(num)
print(type(num))

price = "199.99"

price = float(price)

print(price)
print(type(price))

age = 21

age = str(age)

print(age)
print(type(age))
```
