```python

# What are Tuples?


# A tuple is an ordered and immutable collection.
# Unlike lists, tuples cannot be modified after creation.

numbers = (10, 20, 30, 40)

print(numbers)
print(type(numbers))
```

```python

# Creating Tuples

#no mandatory of brackets you can use parentheses 
# Tuples can store different data types.

numbers = (10, 20, 30)

names = ("Ratan", "Rahul", "Amit")

mixed = (10, "Python", 3.14, True)

nested = ((1, 2), (3, 4))

print(numbers)
print(names)
print(mixed)
print(nested)
```

```python

# Single Element Tuple


# A comma is required when creating a tuple
# with only one element.

num = (10,)

print(type(num))

not_tuple = (10)

print(type(not_tuple))
```

```python

# Packing


# Packing means storing multiple values
# into a single tuple.

student = ("Ratan", 22, "Python")

print(student)
```

```python

# Unpacking


# Unpacking assigns tuple elements
# into separate variables.

name, age, course = ("Ratan", 22, "Python")

print(name)

print(age)

print(course)
```

```python

# Indexing


# Indexing accesses individual tuple elements.

fruits = ("Apple", "Banana", "Mango", "Orange")

print(fruits[0])

print(fruits[2])

print(fruits[-1])
```

```python

# Slicing


# Slicing extracts a portion of a tuple.

numbers = (10,20,30,40,50,60)

print(numbers[1:4])

print(numbers[:3])

print(numbers[3:])

print(numbers[::-1])
```

```python

# count()


# count() returns the number of times
# an element appears in the tuple.

numbers = (10,20,20,30,20)

print(numbers.count(20))
```

```python

# index()


# index() returns the first occurrence
# of an element.

fruits = ("Apple","Banana","Mango")

print(fruits.index("Banana"))
```

```python

# len()


# len() returns the total number of elements.

numbers = (10,20,30,40)

print(len(numbers))
```

```python

# max()


# max() returns the largest element.

numbers = (10,50,30)

print(max(numbers))
```

```python

# min()


# min() returns the smallest element.

numbers = (10,50,30)

print(min(numbers))
```

```python

# sum()


# sum() returns the sum of all numeric elements.

numbers = (10,20,30)

print(sum(numbers))
```

```python

# Find the Second Largest Element


numbers = (12, 45, 7, 89, 34, 89)

unique_numbers = sorted(set(numbers), reverse=True)

print("Second Largest:", unique_numbers[1])
```

```python

# Remove Duplicate Elements


numbers = (10, 20, 30, 20, 40, 10, 50)

result = tuple(dict.fromkeys(numbers))

print(result)
```

```python

# Check Palindrome Tuple


numbers = (1, 2, 3, 2, 1)

if numbers == numbers[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python

# Frequency of Each Element


numbers = (10, 20, 20, 30, 10, 40)

frequency = {}

for num in numbers:
    frequency[num] = frequency.get(num, 0) + 1

print(frequency)
```

```python

# Merge and Sort Tuples


tuple1 = (30, 10, 40)

tuple2 = (20, 60, 50)

result = tuple(sorted(tuple1 + tuple2))

print(result)
```

```python

# Common Elements


tuple1 = (10, 20, 30, 40)

tuple2 = (20, 30, 50, 60)

common = tuple(set(tuple1) & set(tuple2))

print(common)
```

```python

# Rotate Tuple by N Positions


numbers = (10, 20, 30, 40, 50)

n = 2

rotated = numbers[n:] + numbers[:n]

print(rotated)
```

```python

# Student Record Analysis


students = (
    ("Ratan", 85),
    ("Rahul", 92),
    ("Amit", 78),
    ("Priya", 95)
)

topper = students[0]

for student in students:
    if student[1] > topper[1]:
        topper = student

print("Topper:", topper[0])
print("Marks :", topper[1])
```

```python
#using functions

def get_marks(student):
    return student[1]

topper = max(students, key=get_marks)
print("Topper:", topper[0])
print("Marks :", topper[1])


#topper = max(students, key=lambda student: student[1])     we can use also this method
```
