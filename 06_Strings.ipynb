```python
a = "hello"
b = "Hello123"
c = "123456"
d = "HELLO"
e = " "
f = "Hello 123"
g = "1.234"
#isalnum - Returns True if all characters in the string are alphanumeric
#print(a,a.isalnum())
#print(b,b.isalnum())
#print(c,c.isalnum())
#print(d,d.isalnum())
#print(e,e.isalnum())
#print(f,f.isalnum())
#print(g,g.isalnum())
#isalpha - Returns True if all characters in the string are in the alphabet
print(a,a.isalpha())
print(b,b.isalpha())

#isdecimal - Returns True if all characters in the string are decimals
print(c,c.isdecimal())
print(g,g.isdecimal())
#isdigit - Returns True if all characters in the string are digits
print(c,c.isdigit())
#isnumeric - Returns True if all characters in the string are numeric
print(c, c.isnumeric())
print(g, g.isnumeric())

#islower - Converts a string into lower case
print(a, a.islower())
print(d, d.islower())
#isupper - Returns True if all characters in the string are upper case
print(a, a.isupper())
print(d, d.isupper())
#isspace - Returns True if all characters in the string are whitespaces
print(e, e.isspace())
print(a, a.isspace())
#istitle - Returns True if the string follows the rules of a title
print("Hello World", "Hello World".istitle())
print(f, f.istitle())
print(a, a.istitle())
```

```python
#endswith()-Returns true if the string ends with the specified value
a="Harry potter"
print(a.endswith("r"))
```

```python
#startswith()-returns true if the string with the specified value
a="Harry potter"
print(a.startswith("H"))
```

```python
#swapcase()-it changes lower to upper and upper to lower
a="Harry potter"
print(a.swapcase())
```

```python
#strip()- returns a trimmed version of the string
a="&&&&&&Harry potter%%%%%"
print(a.strip("%,&"))
```

```python
#split()-splits the string at the specified separator and returns a list
a="ratan#jivan#bhidhu#"
print(a.split("#"))
```

```python
#ljust()-returns a left justified version of the string
a="harry potter"
x=a.ljust(20,"#")
print(x)
#rjust()-same like ljust but it returns right justified
```

```python
#replace()-returns a string where a specified value is replaced with a specified value
a="my name is john.john love to play"
print(a)
print(a.replace("john","rioc"))
```

```python
#rindex()- searches the string for a specified value and returns the last position of where it was found
a="Harry potter and the prisoner of azkaban"
print(a.rindex("potter"))
```

```python
#rfind()-similar to the rindex()
a="Harry potter and the prisoner of azkaban"
print(a.rfind("potter"))
```

```python
# Write a program to check if a string is a palindrome.

a=input("enter any thing here")
rev=a[::-1]
if rev==a:
    print("palindrome")
else:
    print("not palindrome")
```

```python
# Write a program to find the number of vowels in a string.

a=input("enter any thing here")
vowels=0
for i in a:
    if i=="a"or i=="e" or i=="i"or i=="o"or i=="u":
        vowels+=1
print(vowels)
```
